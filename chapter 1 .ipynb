```python
# // EXERCISE 2 //

def m_d(G):
    f=0
    for n in G.nodes:
        f=max(f,G.degree(n))
    return f
m=m_d(g)
for n in g.nodes:
    if m==g.degree(n):
        s=(n,m)
s
```

```python
g = nx.Graph()
g.add_node('a')
nodes_to_add = ['b', 'c', 'd']
g.add_nodes_from(nodes_to_add)
g.add_edge('a', 'b')
edges_to_add = [('a', 'c'), ('b', 'c'), ('c', 'd')]
g.add_edges_from(edges_to_add)
nx.draw(g, with_labels=True)
```

```python
D = nx.DiGraph()

D.add_edges_from([(1,2),(2,3),(3,2),(3,4),(3,5),(4,5),(4,6),(5,6),(6,4),(4,2)])

nx.draw(D, with_labels=True)
```
